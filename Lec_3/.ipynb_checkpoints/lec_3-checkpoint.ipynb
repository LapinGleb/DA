```python
import requests
import xmltodict
import json
import pandas as pd
```

### Выберите валюту и даты

```python
currency = 'EUR/RUB'
moment_start = '2021-01-01'
moment_end = '2022-01-01'
```

```python
user_agent = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.124 YaBrowser/22.9.2.1495 Yowser/2.5 Safari/537.36'}
r = requests.get('https://www.moex.com/export/derivatives/currency-rate.aspx?'
                 'language=ru&currency={0}&moment_start={1}&moment_end={2}'.format(currency, moment_start, moment_end), 
                 headers=user_agent)
```

```python
xpars = xmltodict.parse(r.text)
j = json.dumps(xpars)
```

```python
d = json.loads(j)
```

```python
df = pd.DataFrame(d.get('rtsdata').get('rates').get('rate'))
```

```python
col = ['date', 'value']
df.columns = col
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date']).dt.date.astype('datetime64[ns]')
```

```python
df['value'] = df.value.astype(float)
```

```python
df = df.sort_values(by='date').reset_index(drop=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
df.set_index('date').plot()
```

```python
df = df.groupby('date')['value'].mean().reset_index()
```

```python
df.set_index('date').plot()
```

```python
df.set_index('date', inplace=True)
```

```python
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['is_weekend'] = data.index.isin([5, 6]) * 1
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['value'].shift(lag)
    
    data['y_mean'] = data['value'].shift().rolling(rolling_mean_size).mean().copy()
```

```python
df
```

```python
make_features(df, 30, 3)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('value', axis=1),
                                                   df.value,
                                                   shuffle=False,
                                                   test_size=0.19)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print('Ошибка на Test', mean_absolute_error(y_test, lr.predict(X_test)))
```

```python
print('Ошибка на Train', mean_absolute_error(y_train, lr.predict(X_train)))
```
